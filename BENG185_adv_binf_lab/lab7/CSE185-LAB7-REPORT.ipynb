## Introduction (10 pts)

In your own words, summarize background and motivation, and major conclusions of your phylogenetic analysis of coronavirus genomes

I tried to reproduce the results of a study that analysed novel coronavirus strains and compare them to other known coronaviruses by creating a phylogenetic tree that can show which viruses are evolutionarily similar. From my analysis, it seems that the novel coronavirus is most similar to Bat SARS-like coronaviruses. We can also see that the novel coronavirus strains and the SARS coronaviruses cluster together, which is a nice sanity check.

## Methods (40 pts)

Describe the methods you used in your analysis. In particular, your methods writeup should include:
* Where did you obtain genome assemblies from? What types of virus strains do they come from?
* What tool did you use for multiple sequence alignment? What version? Did you use any non-default parameters?
* What tool did you use for tree building? What version? Did you use any non-default parameters? Give a better description than the NEJM paper did!

I retrieved the genomes from NCBI using accession numbers by using a quick command line for loop to wget the assemblies and concatenate them into a single file. All of these genomes come from various coronaviruses.

I used mafft (v7.475) with the FFT-NS-2 strategy to do the multiple sequence alignment. Otherwise no other special parameters were used.

I used RaxML (8.2.12) for tree building. I used the seed (-p) 12345 when building 20 maximum likelihood trees and the model of nucleotide substitution (-m) "GTRGAMMA". I then did 100 bootstrap searches (using same params as before). The last step was to draw bipartitions based on bootstrap values. 

To visualize the tree, I used iTOL web version.
